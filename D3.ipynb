```python
import matplotlib.pyplot as plt

x = ['a', 'b', 'c']
y = [9, 2, 7]
plt.bar(x, y, color=['red', 'green', 'blue'])
plt.title("Bar Chart")
plt.show()

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
plt.scatter(x, y, color='purple', marker='o')
plt.title("Scatter Plot")
plt.show()

data = [1, 2, 2, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins=4, color='orange', edgecolor='black')
plt.title("Histogram")
plt.show()

labels = ["apple", "banana", "grapes"]
sizes = [50, 30, 20]
colors = ['#ff9999', '#66b3ff', '#99ff99']
plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors)
plt.title("Pie Chart")
plt.show()
```
